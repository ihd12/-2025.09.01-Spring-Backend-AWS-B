```python
# 파일 객체 생성
# open("파일경로","모드")
# 모드의 종류 
# w(쓰기모드) : 파일이 없는 경우 생성, 파일이 있는 경우 지우고 생성
# a(추가모드) : 파일이 없는 경우 생성, 파일이 있으면 원래 내용에 추가
# r(읽기모드) : 파일의 내용을 읽는 용도로 사용
f = open("C:/python/새파일.txt", "w")
# 파일 닫기
f.close()
```

```python
# 파일에 내용 추가하기 w모드
f = open("C:/python/새파일.txt", "w")
# write(파일에 추가할 내용)
f.write("안녕하세요\n")
for i in range(1,6):
    f.write(f"{i}번째 줄입니다.\n")
f.close()
```

```python
# 파일에 내용 추가하기 a모드
f = open("C:/python/새파일.txt", "a")
# write(파일에 추가할 내용)
f.write("안녕하세요\n")
for i in range(6,11):
    f.write(f"{i}번째 추가 줄입니다.\n")
f.close()
```

```python
# 한줄씩 읽기
f = open("C:/python/새파일.txt", "r")
while True:
    # 파일에서 한줄을 읽고 line변수에 저장
    line = f.readline()
    # line 변수에 데이터가 없으면 break실행
    if not line : break
    # line 변수에 데이터가 있으면 출력
    print(line, end="")
f.close()
```

```python
# 파일에서 모든 줄을 읽어 리스트로 반환하는 방식
f = open("C:/python/새파일.txt", "r")
lines = f.readlines()
print(lines)
f.close()
for i in lines:
    print(i, end="")
```

```python
# 파일의 전체 내용을 읽고 문자열로 반환
f = open("C:/python/새파일.txt", "r")
data = f.read()
print(data)
f.close()
```

```python
# with open : 파일사용후 자동으로 닫는 기능
with open("C:/python/새파일.txt",'r') as f:
    data = f.read()
    print(data)
# 파일이 닫힌 후에는 f를 사용할 수 없음
# data2 = f.read() # 에러 발생
```

```python
# 구구단 2~9단까지의 내용을 파일에 저장하기
# 파일 이름은 gugudan.txt
# 출력 내용 : 
# 2*1=2, 2*3=6, 2*4=8 ....
# 3*1=3, 3*2=6, 3*3=9 ....
with open("c:/python/gugudan.txt", "w") as f:
    for i in range(2,9+1):
        for j in range(1,9+1):
            f.write(f"{i}*{j}={i*j}, ")
        f.write("\n")
```
