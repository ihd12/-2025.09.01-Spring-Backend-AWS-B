```python
# if문
if(True):
    # 실행문의 경우 들여쓰기 모두 같아야 정상적으로 실행됨
    print("조건식이 True의 경우 실행")
    print("두번째 실행문")
# 들여쓰기를 하지 않으면 if문이 끝난것으로 설정됨
print("if문이 끝났습니다.")

# pass : if문 안에 실행문이 없이 실행하도록 만드는 명령어
if(True):
    pass
print("if문 종료")
```

```python
# if, elif, else
a = 10
if(a>10):
 print("a는 10보다 큽니다.")
elif(a<10):
  print("a는 10보다 작습니다.")
  print("두번째 실행문")
else:
   print("a는 10과 같습니다.")
```

```python
# 논리 연산자
# >, >=, <, <=, ==, !=, and, or
print("10!=10 ", 10!=10)
print("10==10 ",10==10)
print("15>=10 ",15>=10)
print("15>10 ",15>10)
print("15<15 ",15<15)
print("15<=15 ",15<=15)
print("3>5 and 5>3 ",3>5 and 5>3)
print("3>5 or 5>3 ",3>5 or 5>3)
```

```python

#문자 자료형 비교
a = '홍길동'
b = '이순신'
c = '홍길동'
print(a==b)
print(a==c)
print(a!=c)
a = 'a'
b = 'b'
print(a>b) # A~z 사전순 비교 
print(a<b) # A가 가장 작고 z가 가장 크다
print('홍길동'>'이순신')
print('홍길동'<'이순신')
# is 연산자 : 객체 비교 연산자
a1 = 'abc'
b2 = 'abc'
c3 = a1
print(a1==b2) # 문자열끼리 비교
print(a1==c3) 
print(a1 is b2) # 객체끼리 비교 -> 결과가 true 때도 false일때도 있음
print(a1 is c3)
```

```python
# in, not in 연산자
a = [1,3,5,7,9]
print(2 in a) # a리스트 안에 1이 있으면 true 없으면 false
print(2 not in a) # a리스트 안에 1이 있으면 false 없으면 true
b = 'python is fun'
print('fun' in b) # b문자열 안에 fun이 있으면 true 없으면 false
print('fun' not in b) # b문자열 안에 fun이 있으면 false 없으면 true
```
