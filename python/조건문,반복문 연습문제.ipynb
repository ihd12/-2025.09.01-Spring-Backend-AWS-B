```python
# 0 : 구구단 2~9단까지 출력하기
for i in range(2,9+1):
    for j in range(1,9+1):
        print(i*j, end=" ")
    print()
```

```python
# 1 : while문을 사용해 1부터 1000까지의 자연수 중 3의 배수의 합을 구해 보자.
i = 1
total = 0
while(i<=1000):
    if(i%3==0):
        total+=i
    i+=1
print(total)
```

```python
# 2 : a 리스트의 총합을 구해 보자.
a = [70,60,55,75,95,90,80,80,85,100]
total = 0
for i in a:
    total+=i
print(total)
print(sum(a))
```

```python
# 3 : 1~100 중 홀수만을 저장하는 리스트를 만들어 보자
a = []
for i in range(1,100+1):
    if(i%2!=0):
        a.append(i)
print(a)
#리스트 내포 : 리스트를 생성하기 편하게 한줄로 작성하는 방식
# [표현식 for 변수 in 반복가능한객체 if 조건]
b = [i+1 for i in range(1,100+1) if(i%2!=0)]
print(b)
```

```python
# 4 : 리스트에서 글자 수가 세글자 이상인 문자만 출력해보자
a = ['Life','is','too','short',',','you','need','python']
for i in a:
    if(len(i)>=3):
        print(i, end=" ")
```

```python
# 5 : 1900~2023년 까지의 윤년을 출력해보자]
for i in range(1900, 2023+1):
    if(i%400==0 or (i%100!=0 and i%4==0)):
        print(i , end=" ")
```

```python
# 6 : 가위바위보 게임에서 승리 횟수가 5회 이상일때 게임이 끝나도록 만들어 보자
# 출력결과 : '승리 : 0/5', 패배 : 1, 비김 : 0'
#'승리 : 1/5', 패배 : 1, 비김 : 0'
#'승리 : 2/5', 패배 : 1, 비김 : 0'
#'승리 : 3/5', 패배 : 1, 비김 : 0'
#'승리 : 3/5', 패배 : 2, 비김 : 0'
#'승리 : 4/5', 패배 : 2, 비김 : 0'
#'승리 : 5/5', 패배 : 2, 비김 : 0'
#'다섯번 이겼습니다. 축하합니다.'
import random
data = ['가위','바위','보']
win = 0
lose = 0
draw = 0
while(win<5):
    computer = random.choice(data)
    player = input("가위,바위,보 중에 입력해주세요>>")
    print("플레이어 :", player)
    print("컴퓨터 : ", computer)
    if(player == computer):
        draw+=1
    elif(player=="가위" and computer=="보" 
        or player=="바위" and computer=="가위" 
        or player=="보" and computer=="바위"):
        win+=1
    else:
        lose+=1
    # f-string 출력방식 : 문자열 안에 변수를 적기 쉽도록 하는 출력방식
    print(f'승리 : {win}/5, 패배 : {lose}, 비김 : {draw}')
print('다섯번 이겼습니다. 축하합니다.')
```
