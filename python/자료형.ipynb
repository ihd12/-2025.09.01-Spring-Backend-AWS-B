```python
# 파이썬은 #을 주석 기호로 사용
# 숫자 자료형 : 정수, 실수, 복소수, 8진수, 16진수
# 파이썬의 경우 저장하는 데이터의 형식에 따라 자료형이 설정됨
a = 100 # a변수에 100을 저장 => 정수
b = 10.5 # b변수에 10.5를 저장 => 실수
c = 0o12 # c변수에 8진수 12를 저장 => 8진수
d = 0xA # d변수에 16진수 A를 저장 => 16진수
print(a)
print(b)
# 8진수, 16진수는 10진수로 변경하여 출력됨
print(c)
print(d)
print(a+b+c+d) # 10진수 실수형으로 출력됨
```

```python
# 숫자 연산자
# 사칙연산 : +,-,*,/, %(나머지 연산)
# 제곱연산자 : **
print(2**2)
print(2**32)
print(2**64)
# 소수점을 버리는 나누기 연산자, 몫만 출력하는 연산자 : //
print(7/3)
print(7//3)
```

```python
# 문자열 자료형 : 큰따움표,작은따움표 상관없이 문자열을 저장할 수 있음.
# 역슬래시(\)를 사용하여 문자열안에 따움표를 문자로써 사용 가능
a = "'큰따움표' 사용 \" "
b = '"작은따움표" 사용'
# 문자열의 경우 엔터 사용시 에러가 나지 않도록 하는 문자열 작성 방식
c = """큰따움표 
3개 사용"""
d = '''작은 따움표 3개 
사용'''
print(c)
print(d)
```

```python
#문자열 슬라이싱 : 특정 문자열을 잘라내어 출력하는 방식
a = "python is fun!"
print(a[5:7+1])
print(a[7:8+1])
# 왼쪽을 생략하는 슬라이싱
print(a[:7]) # 처음부터 7번 문자열까지 출력
print(a[:-1]) # 처음부터 뒤에서 1번 문자열까지 출력 
# 오른쪽을 생략하는 슬라이싱
print(a[7:]) # 7번부터 마지막까지 출력
print(a[-5:]) # 뒤에서 5번째 부터 마지막까지 출력
```
