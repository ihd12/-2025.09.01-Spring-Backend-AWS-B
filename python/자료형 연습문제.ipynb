```python
#주어진 실수 3.1415를 소수점 두 번째 자리에서 반올림하는 파이썬 코드를 작성하세요.
a=3.1415
print(round(a)) # 소수점 첫번째 자리에서 반올림
print(round(a,0))
print(round(a,1)) # 소수점 두번째에서 반올림
print(round(a,2)) # 소수점 세번째에서 반올림
b = 4365.715
print(round(b,-1)) # 정수 첫번째에서 반올림 
print(round(b,-2)) # 정수 두번째에서 반올림
```

```python
#주어진 리스트 [1, 2, 3, 4, 5]에서 첫 번째 원소와 마지막 원소를 교환하는 파이썬 코드를 작성하세요.
#자바, 파이썬 둘다 가능한 방식
a = [1, 2, 3, 4, 5]
temp = a[0]
a[0] = a[4]
a[4] = temp
print(a)
# 파이썬에서만 가능한 방식
a = [1, 2, 3, 4, 5]
a[0],a[4] = a[4],a[0]
print(a)
```

```python
#주어진 리스트 [1,3,5,4,2]를 내림차순으로 정렬하는 파이썬 코드를 작성하세요.
a=[1,3,5,4,2]
a.sort(reverse=True)
print(a)
```
